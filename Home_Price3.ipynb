```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_excel('Home_Price3.xlsx')
df
```

```python
dummies=pd.get_dummies(df.Town)
```

```python
merged=pd.concat([df,dummies],axis="columns")
merged
```

```python
final=merged.drop(['Town','West Winder'],axis='columns')
final
```

```python
model=LinearRegression()
```

```python
x=final.drop('Price',axis='columns')
x
```

```python
y=final.Price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])  # price of a home in Robinsville of size 2800
```

```python
model.predict([[3400,0,0]])  # price of a home in west Winder of size 3400  
```

```python
model.score(x,y)
```
